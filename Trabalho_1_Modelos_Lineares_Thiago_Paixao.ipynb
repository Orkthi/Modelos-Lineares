```python
#Questão d

x_barra = sum(X) / len(X)
y_barra = sum(Y) / len(Y)

numerador = 0
denominador_x = 0
denominador_y = 0

for xi, yi in zip(X, Y):
    numerador += (xi - x_barra) * (yi - y_barra)
    denominador_x += (xi - x_barra)**2
    denominador_y += (yi - y_barra)**2

coef = numerador / ( (denominador_x * denominador_y)**0.5 )

print
coef
```

```python
soma_x = sum(X)
soma_y = sum(Y)
soma_x2 = sum(xi**2 for xi in X)
soma_xy= sum(xi*yi for xi, yi in zip(X, Y))
n = len(X)

a = (n*soma_xy - soma_x*soma_y) / (n*soma_x2 - soma_x**2)
b = (soma_y*soma_x2 - soma_x*soma_xy) / (n*soma_x2 - soma_x**2)


variancia = sum((yi - a*xi - b)**2 for xi, yi in zip(X, Y)) / n
desvio = variancia**0.5

print(f'Coeficiente angular (a): {a}')
print(f'Coeficiente linear (b): {b}')
print(f'Desvio padrão dos resíduos: {desvio}')

reta = [a*xi + b for xi in X]


plt.scatter(X, Y, color='blue', label='Pontos de Dados')
plt.plot(X, reta, color='red')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Reta dos mínimos quadrados')
plt.show()
```
