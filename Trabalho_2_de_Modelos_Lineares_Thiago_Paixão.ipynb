b) Estime a correlação entre a variável Y e cada uma das variáveis regressoras (ρ(Y,Xi)).

```python
def calcular_correlacoes(Y, *Xs):
    correlacoes = {}
    # Calcular a média de Y e X
    media_Y = np.mean(Y)
    media_X = np.mean(Xs[0])  # Use Xs[0] para calcular a média de X1
    for i, X in enumerate(Xs, start=1):
        # Calcular os desvios em relação à média
        desvio_Y = Y - media_Y
        desvio_X = X - media_X

        # Calcular o produto dos desvios
        produto_desvios = desvio_Y * desvio_X

        # Calcular a correlação de Pearson
        correlacao = np.sum(produto_desvios) / (np.sqrt(np.sum(desvio_Y**2)) * np.sqrt(np.sum(desvio_X**2)))

        correlacoes[f'ρ(Y,X{i})'] = correlacao


    return correlacoes

# Imprimir o resultado
correlacoes = calcular_correlacoes(Y, X1, X2)
for key, value in correlacoes.items():
    print(f'{key}: {value}')
```

Os valores da correlação foram bem próximo de 1 pois, as variáveis "Roubos em coletivo" e "Roubos a transeunte" apresentam comportamento de forma diretamente proporcional à variável "Total de roubos", mostrando que o número de roubos em coletivos e de roubos a transeunte contribuem significativamente para o número total de roubos.

d) Encontre o hiperplano de quadrados mínimos (estime β0, β1, ..., βp).

```python
def encontrar_hiperplano(X1, X2, Y):
    # Construir a matriz X
    X = np.column_stack((np.ones_like(X1), X1, X2))

    # Calcular os coeficientes B
    Xt = X.T
    XtX_inv = np.linalg.inv(np.dot(Xt, X))
    beta = np.dot(np.dot(XtX_inv, Xt), Y)

    return beta

coeficientes = encontrar_hiperplano(X1, X2, Y)
print(coeficientes)
```

e) Calcule os valores estimados de y (ou seja, ŷ) através do hiperplano estimado no item anterior.

```python
def calcular_valores_estimados(X1, X2, beta):
    # Construir a matriz X
    X = np.column_stack((np.ones_like(X1), X1, X2))

    # Calcular ŷ usando o hiperplano estimado
    y_estimado = np.dot(X, beta)

    return y_estimado

coeficientes = encontrar_hiperplano(X1, X2, Y)  # Hiperplano estimado
y_estimado = calcular_valores_estimados(X1, X2, coeficientes)  # Valores estimados de y

# Imprimir os valores estimados de y
print(y_estimado)
```

f) Calcule os resíduos, ou seja, o erro de estimação y-ŷ.

```python
residuos = Y - y_estimado

# Exibindo os resíduos
print(residuos)
```
